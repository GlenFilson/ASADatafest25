# Load in the datasets
We are going to combine the US Zip Code dataset with the Leases dataset
This will help us get leasing as latitude and longitude so we can plot it on a map for visualisation

```python
leases = pd.read_csv('../Datasets/Leases.csv')
zip_codes = pd.read_csv('../Datasets/uszips.csv')

leases.head()
```

# Merging the two tables
So we can access latitude and longitude 

```python
result = pd.merge(
    leases,
    zip_codes[['zip', 'lat', 'lng']],
    on='zip',
    how='left'
)
result = result.dropna(subset=['lat', 'lng'])
len(result)
```
